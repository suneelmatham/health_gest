```python
df_copy = df.copy()
```

```python
train = df[df.data == 'train']
test = df[df.data == 'test']
```

```python
len(test)
```

```python
y_train = train.GA
y_test = test.GA
```

```python
X_train = train.drop(['GA','time','randPerson','SampleID','data'], axis=1)
X_test = test.drop(['GA','time','randPerson','SampleID','data'], axis=1)
```

```python
X_test.head()
```

```python
# Feature scaling
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler

num_pipeline = Pipeline([('min_max_scaler', MinMaxScaler())])
X_train_tr = num_pipeline.fit_transform(X_train)
```

## Random Forest

```python
# Random forest
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(n_estimators=100 ,random_state=1)
rf_model.fit(X_train_tr, y_train)
```

```python
# training error
from sklearn.metrics import mean_absolute_error

preds_train = rf_model.predict(X_train_tr)
mean_absolute_error(y_train, preds_train)
```

```python
# Cross Validation
from sklearn.model_selection import cross_val_score
scores = cross_val_score(rf_model, X_train_tr, y_train, 
                         scoring ='neg_mean_absolute_error', cv=10)
def display_scores(scores):
    print('Scores:', scores)
    print('Mean: ', scores.mean())
    print('Standard Deviation:', scores.std())
    
display_scores(-scores)
```

```python
# Saving model
import joblib

joblib.dump(rf_model, 'rf_model.pkl')
```

## Elastic Net

```python
from sklearn.linear_model import ElasticNet
elastic_net = ElasticNet(alpha=0.1, l1_ratio=0.5)
elastic_net.fit(X_train_tr, y_train)
```

```python
# Training error
preds_train = elastic_net.predict(X_train_tr)
mean_absolute_error(y_train, preds_train)
```

```python
scores = cross_val_score(elastic_net, X_train_tr, y_train, 
                         scoring ='neg_mean_absolute_error', cv=10)
display_scores(-scores)
```

```python
joblib.dump(elastic_net, 'elastic_net.pkl')
```

## Test Set

```python
X_test_tr = num_pipeline.transform(X_test)
preds = rf_model.predict(X_test_tr)
```

```python
# testing error
mean_absolute_error(y_test, preds)
```
