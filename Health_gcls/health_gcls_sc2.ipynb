```python
df_copy = df.copy()
```

```python
train = df[df.data == 'train']
test = df[df.data == 'test']
```

```python
len(test)
```

```python
y_train = train.GA
y_test = test.GA
```

```python
X_train = train.drop(['GA','time','randPerson','SampleID','data'], axis=1)
X_test = test.drop(['GA','time','randPerson','SampleID','data'], axis=1)
```

```python
X_test.head()
```

```python
# Feature scaling
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler

num_pipeline = Pipeline([('min_max_scaler', MinMaxScaler())])
X_train_tr = num_pipeline.fit_transform(X_train)
```

## Random Forest

```python
# Random forest
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(n_estimators=100 ,random_state=1)
rf_model.fit(X_train_tr, y_train)
```

```python
# training error
from sklearn.metrics import mean_absolute_error

preds_train = rf_model.predict(X_train_tr)
rf_train_err = mean_absolute_error(y_train, preds_train)
rf_train_err
```

```python
# Cross Validation
from sklearn.model_selection import cross_val_score
scores = cross_val_score(rf_model, X_train_tr, y_train, 
                         scoring ='neg_mean_absolute_error', cv=10)
def display_scores(scores):
    print('Scores:', scores)
    print('Mean: ', scores.mean())
    print('Standard Deviation:', scores.std())
    
display_scores(-scores)
```

```python
rf_valid_err = -scores.mean()
rf_valid_err
```

```python
# Saving model
import joblib

joblib.dump(rf_model, 'rf_model.pkl')
```

```python
from sklearn.model_selection import GridSearchCV
# 'max_features':[2,4,6,8]}
param_grid = [{'n_estimators':[10,100,1000]},
             {'bootstrap':[False],'n_estimators':[10,100,1000]}]
forest_reg = RandomForestRegressor()
grid_search = GridSearchCV(forest_reg, param_grid, cv=10, scoring ='neg_mean_absolute_error'
                          , return_train_score = True)
grid_search.fit(X_train_tr, y_train)
```

```python
grid_search.best_params_
```

```python
cvres = grid_search.cv_results_
for mean_score, params in zip(cvres['mean_test_score'], cvres['params']):
    print(-mean_score, params)
```

## Elastic Net

```python
from sklearn.linear_model import ElasticNet
elastic_net = ElasticNet(alpha=0.1, l1_ratio=0.5)
elastic_net.fit(X_train_tr, y_train)
```

```python
# Training error
preds_train = elastic_net.predict(X_train_tr)
mean_absolute_error(y_train, preds_train)
```

```python
scores = cross_val_score(elastic_net, X_train_tr, y_train, 
                         scoring ='neg_mean_absolute_error', cv=10)
display_scores(-scores)
```

```python
joblib.dump(elastic_net, 'elastic_net.pkl')
```

```python
param_grid = [{'alpha':[0.1,0.5,0.8], 'l1_ratio':[0.1,0.5,0.8]}]
elastic_reg = ElasticNet()
grid_search = GridSearchCV(elastic_reg, param_grid, cv=10, scoring ='neg_mean_absolute_error'
                          , return_train_score = True)
enet_model = grid_search.fit(X_train_tr, y_train)
```

```python
grid_search.best_params_
```

```python
cvres = grid_search.cv_results_
for mean_score, params in zip(cvres['mean_test_score'], cvres['params']):
    print(-mean_score, params)
```

```python
y_preds = enet_model.predict(X_train_tr)
enet_train_err = mean_absolute_error(y_train, y_preds)
enet_train_err
```

```python
enet_valid_err = min(-cvres['mean_test_score'])
enet_valid_err
```

## Neural net

```python
import tensorflow as tf
from tensorflow import keras
keras.__version__
```

```python
tf.__version__
```

```python
X_train_tr.shape[1:]
```

```python
from sklearn.model_selection import train_test_split

X_train_tr1, X_valid, y_train1, y_valid = train_test_split(X_train_tr, y_train )
```

```python
model = keras.models.Sequential([keras.layers.Dense(300, activation='relu',
                                                   input_shape=X_train_tr.shape[1:]),
                                  keras.layers.Dense(150, activation ='relu'),
                                keras.layers.Dense(1)])
model.compile(loss='mean_absolute_error', optimizer='sgd')
history = model.fit(X_train_tr1, y_train1, epochs=20, validation_data = [X_valid, y_valid])
```

```python
history = model.fit(X_train_tr1, y_train1, epochs=20, validation_data = [X_valid, y_valid])
```

```python
import matplotlib.pyplot as plt

pd.DataFrame(history.history).plot(figsize=(8,5))
plt.grid(True)
# plt.gca().set_ylim(0,1)
plt.show()
```

```python
mlp_train_err = model.evaluate(X_train_tr, y_train)
mlp_train_err
```

```python
mlp_valid_err = model.evaluate(X_valid, y_valid)
mlp_valid_err
```
